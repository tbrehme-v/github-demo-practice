# Simple Data Analysis Demo

This notebook demonstrates basic data analysis on sample data

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('data/sample_data.csv')
print(df.head())
```

```python
# Basic statistics
print(df.describe())
print(f"/nAverage score: {df['score'].mean():.2f}")
```

```python
# Simple visualization
df.groupby('category')['score'].mean().plot(kind='bar')
plt.title('Average Score by Category')
plt.ylabel('Score')
plt.show()
```

```python
# Distribution of ages
plt.figure(figsize=(8, 5))
plt.hist(df['age'], bins=8, edgecolor='black', color='skyblue')
plt.title('Distribution of Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.5)
plt.show()
```

```python
# Scatter plot: Age vs Score
plt.figure(figsize=(8, 5))
plt.scatter(df['age'], df['score'], c=df['category'].map({'A': 'blue', 'B': 'red'}), 
    s=100, alpha=0.6, edgecolor='black')
plt.title('Age vs Score by Category')
plt.xlabel('Age')
plt.ylabel('Score')
plt.legend(['Category A', 'Category B'])
plt.grid(True, alpha=0.3)
plt.show()
```
